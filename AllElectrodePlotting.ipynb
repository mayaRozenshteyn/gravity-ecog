```python
from nilearn import plotting
import scipy.io
import matplotlib.pyplot as plt
```

```python
id_eleccount = # Removed to protect patient anonymity
```

```python
def plot_elec_on_brain():
    
    patient_list = []
    elecpost_list = []
    
    i = 1
    
    for key in id_eleccount.keys():
        elecpos = scipy.io.loadmat('/scratch/gpfs/mayaar/GravityECoG/derivatives/preprocessing/sub-ny' + key + '/eeg1_manualica_notch_elec_elecpos.mat')['trial']
        
        for j in range(id_eleccount[key]):
            elecpost_list.append(elecpos[j, :])
            patient_list.append(i)
            
        i += 1
            
    fig = plt.figure(figsize=(12, 6))
    plotting.plot_markers(patient_list,
                    elecpost_list,
                    node_size=20,
                    node_vmin=1,
                    node_vmax=9,
                    node_cmap=plt.cm.jet,
                    node_threshold=None,
                    display_mode='lzry',
                    figure=fig,
                    axes=None,
                    title=None,
                    annotate=True,
                    colorbar=True)

    return fig
```

```python
fig = plot_elec_on_brain()
```
